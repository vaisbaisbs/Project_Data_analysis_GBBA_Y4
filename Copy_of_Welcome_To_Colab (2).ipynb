```
# 1. Charger
df = pd.read_csv('/Users/sachabreugnon/Desktop/pret_defaut/data/application_data_clean.csv',
                 sep=';', header=0)

print(f"üìä Donn√©es: {df.shape[0]:,} lignes √ó {df.shape[1]} colonnes")
print(f"üî§ Noms des colonnes: {list(df.columns)[:10]}...")
```

```
# 2. Nettoie
# Doublons
doublons = df.duplicated().sum()
if doublons > 0:
    df = df.drop_duplicates()
    print(f"‚úÖ {doublons} doublons supprim√©s")
```

```
# Missing values
missing = df.isnull().sum().sum()
print(f"üîç {missing:,} valeurs manquantes")
df_clean = df.dropna()
print(f"Lignes supprim√©es: {len(df) - len(df_clean)}")
```

```
# 3. Analyse de TARGET
if 'TARGET' not in df.columns:
    print(f"\n‚ùå ERREUR: Colonne 'TARGET' non trouv√©e!")
    print(f"   Colonnes disponibles: {list(df.columns)}")

print(f"\n‚úÖ Colonne TARGET trouv√©e!")
print(f"   Type de donn√©es: {df['TARGET'].dtype}")
print(f"   Valeurs uniques: {sorted(df['TARGET'].dropna().unique().tolist())}")
```

```
# 4. Compter positifs et n√©gatifs
cas_positifs = (df['TARGET'] == 1).sum()   # D√©fauts
cas_negatifs = (df['TARGET'] == 0).sum()   # Non-d√©fauts
total = cas_positifs + cas_negatifs

print(f"\n‚úÖ ANALYSE TARGET:")
print("-" * 30)
print(f"Cas positifs (d√©fauts = 1)   : {cas_positifs:,}")
print(f"Cas n√©gatifs (non-d√©fauts = 0): {cas_negatifs:,}")
print(f"Total √©chantillon            : {total:,}")
```

```
# 5. Calculer les proportions
proportion_positifs = (cas_positifs / total) * 100 if total > 0 else 0
proportion_negatifs = (cas_negatifs / total) * 100 if total > 0 else 0

print(f"\nüìà PROPORTIONS:")
print("-" * 30)
print(f"Proportion de d√©fauts      : {proportion_positifs:.2f}%")
print(f"Proportion de non-d√©fauts  : {proportion_negatifs:.2f}%")
```

```
# 6. regarder les types de donn√©es
print("Types de donn√©es par colonne:")
for col in df.columns:
    dtype = df[col].dtype
    print(f"{col}: {dtype}")
```

```
# 7. Sauvegarde
df.to_csv('/Users/sachabreugnon/Desktop/pret_defaut/data/data_clean.csv',
          index=False, sep=';')
print(f"\nüíæ Donn√©es nettoy√©es sauvegard√©es")
```
